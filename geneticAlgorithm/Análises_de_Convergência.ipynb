```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```
scores = pd.read_csv("https://raw.githubusercontent.com/rafaelgdp/smart-pacman/master/geneticAlgorithm/scores_38gen.txt")
scores
```

```
g = ['G'+str(n) for n in range(0,38,1)]

fig, ax = plt.subplots(figsize=(16, 8))
ax = sns.boxplot(x='geration', y='averageScore', data=scores)
ax.set_xticks(np.arange(38) )
ax.set_xticklabels(g)
ax.set_ylabel('Pontuação', fontsize=15)
ax.set_xlabel('Geração', fontsize=15)
ax.set_title('Análise da Convergência do Score Médio')
plt.show()
```
